### Pokemon Data Analytics Practics 

```python
# import requirements
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Fetch all data

pokemon = pd.read_csv("data/pokemon.csv")
pokemon
```

```python
# Here is to execute in a simple bar graph

# pokemon.plot(x="Type 1", y="Total")

plt.figure(figsize=(15, 8))
plt.scatter(x=pokemon["Type 1"], y=pokemon["Total"])
plt.xlabel("Type 1")
plt.ylabel("Total")
plt.grid()
plt.show()
```

```python
# Get the top 10 most pokemon according to the total
top10 = pokemon[pokemon.Total >= 700]
top10_list_unsorted = list(top10.Total)
top10_list_unsorted_names = list(top10.Name)
top10_list_unsorted_ave = list((top10.Attack + top10.Defense) / 2)
```

```python
# Sort
for i in range(len(top10_list_unsorted)):
	for j in range(i):
		if top10_list_unsorted[i] > top10_list_unsorted[j]:
			top10_list_unsorted[i], top10_list_unsorted[j] = top10_list_unsorted[j], top10_list_unsorted[i]
			top10_list_unsorted_names[i], top10_list_unsorted_names[j] = top10_list_unsorted_names[j], top10_list_unsorted_names[i]
			top10_list_unsorted_ave[i], top10_list_unsorted_ave[j] = top10_list_unsorted_ave[j], top10_list_unsorted_ave[i]
			# Resorting the lists based in the average of Attack and Defense
			# But still not accurate
			if top10_list_unsorted[i] == top10_list_unsorted[j]:
				for k in range(len(top10_list_unsorted_ave)):
					for l in range(k):
						if top10_list_unsorted_ave[k] > top10_list_unsorted_ave[l]:
							top10_list_unsorted[k], top10_list_unsorted[l] = top10_list_unsorted[l], top10_list_unsorted[k]
							top10_list_unsorted_names[k], top10_list_unsorted_names[l] = top10_list_unsorted_names[l], top10_list_unsorted_names[k]
							top10_list_unsorted_ave[k], top10_list_unsorted_ave[l] = top10_list_unsorted_ave[l], top10_list_unsorted_ave[k]

print("Total", "Average", "Name")

for i in range(15):
	print(top10_list_unsorted[i], top10_list_unsorted_ave[i], top10_list_unsorted_names[i])
```
